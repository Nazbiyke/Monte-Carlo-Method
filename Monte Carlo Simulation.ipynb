# Area of the circle

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate  
```

```python
square_size = 2
radius = 1
N = 1000000
dot_size = 5
```

```python
points = np.random.uniform(-1, 1 , size = (N, 2))
distances = (points[:,0]**2 + points[:,1]**2)**0.5
inside = distances <= radius 
```

```python
e_area = (np.sum(inside) / N) * (square_size**2)
i_area = integrate.dblquad(lambda theta, r: r, 0, 1, lambda r: 0, lambda r: 2*np.pi)
print(f"Estimated area: {e_area} ")
print(f"Real area: {i_area} ")
```

```python
plt.scatter(points[:, 0], points[:, 1], c=inside, cmap='PiYG', s=dot_size)
circle = plt.Circle((0,0), radius, color = 'red', fill = False)
plt.gca().add_patch(circle)
plt.gca().set_aspect("equal")
plt.title("Monte Carlo Circle Area")
plt.show()
```

# Volume of n-dimensional sphere

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate  
```

```python
square_size = 2
radius = 1
N = 1000000
dot_size = 5
dim = int(input('Enter the dimension: '))
```

```python
points = np.random.uniform(-1, 1 , size = (N, dim))
distances = (np.sum(points**2, axis=1))**0.5
inside = distances <= radius
e_area = (np.sum(inside)/N) * (square_size ** dim)
```

```python
plt.axhline(e_area, color='blue', linestyle='--')
plt.xscale('log')
plt.show()
```

# Volume of the sphere

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
square_size = 2
radius = 1
N = 1000000
dot_size = 5
dim = 3 
points = np.random.uniform(-1, 1, size=(N, dim))
distances = (np.sum(points**2, axis=1))**0.5
inside = distances <= radius
e_area = (np.sum(inside) / N) * (square_size ** dim)
print(f"Estimated area: {e_area}")
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=inside, cmap='PiYG', s=dot_size)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Monte Carlo Circle Area (3D)')

plt.show()
```
